```python
from sklearn import datasets
from sklearn import model_selection
import numpy as np
import pandas as pd
diabetes = datasets.load_diabetes()
X = diabetes['data'][:,0]
Y = diabetes['target'].reshape(-1,1)
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X,Y, test_size = 0.3)
def fit(X_train,Y_train):
    num = (X_train*Y_train).mean() - X_train.mean()*Y_train.mean()
    den = (X_train**2).mean() - X_train.mean()**2
    m = num/den
    c = Y_train.mean() - m*X_train.mean()
    return m,c
def predict(x,m,c):
    return m*x+c
def score(Y_truth,Y_pred):
    u = ((Y_truth - Y_pred)**2).sum()
    v = ((Y_truth - Y_truth.mean())**2).sum()
    return 1 - (u/v)
Y_train = Y_train.reshape(-1,1)
Y_test = Y_test.reshape(-1,1)
m,c = fit(X_train,Y_train) 
Y_pred = predict(X_test,m,c)
score(Y_test,Y_pred)
```

Note: Inside the function "traintestsplit" use the "random_state = 42". To know more about randomstate please read the document given regarding randomstate. Output Print the testing and training scores in separate lines rounded off to 3 decimal places.

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# load data
fuelEconomy = np.loadtxt('FuelEconomy.csv', delimiter=',')
X = fuelEconomy[:, 0].reshape(-1, 1)
y = fuelEconomy[:, 1]

# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# create a Linear Regressor and fit the training data
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# print training and testing scores rounded off to 3 decimal places
train_score = round(regressor.score(X_train, y_train), 3)
test_score = round(regressor.score(X_test, y_test), 3)

print(test_score)
print(train_score)
```
